```python
#trends of songs over time
cols=['danceability','liveness','energy','instrumentalness','loudness','speechiness','valence']
for col in cols:
        y=data.groupby('year')[col].mean()
        x=y.index
        plt.figure(figsize=(12,4))
        sns.set_style('darkgrid')
        sns.lineplot(x,y,data=data)
```

```python
plt.figure(figsize=(16, 8))
sns.set(style="whitegrid")
x = data.groupby("artists")["popularity"].sum().sort_values(ascending=False).head(20)
ax = sns.barplot(x.index, x)
ax.set_title('Top Artists with Popularity by Mean ')
ax.set_ylabel('Popularity')
ax.set_xlabel('Artists')
plt.xticks(rotation = 90)

#popularity of artists by sum
```
